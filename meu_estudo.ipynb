```python
pilha = [1, 2, 3]
pilha.append (4)  # Isso é um 
" Push"
print(pilha)
```

```python
prato_tirado = pilha.pop()
print (" Tirei o prato:", prato_tirado)
print ("Pilha agora:", pilha)

```

```python
# Criando uma fila de pessoas no banco
fila = ["Helena", "João", "Maria"]
# Chegou mais uma pessoa(push no final)
fila.append("Ana")
# A primeira pessoa da fila é atendida(pop do início)
pessoa_atendida = fila.pop(0)
print("Pessoa atendida:", pessoa_atendida)
print("Fila agora:", fila)
```

```python
# Passo 1: Criar a lista(O seu banco de dados )
cadastro_jogadores = []
# Passo 2: O comando de repetição 
while True:
# Passo 3: Capturar os dados (Ouvir o que você digita)
    nome = input("Qual é o nome do amigo? ")
    jogo = input("Qual é o jogo favorito do amigo? ")
# Passo 4:Criar a ficha do jogador 
    jogador = {"nome": nome, "jogo": jogo}
#passo 5 = Guardar a lista
    cadastro_jogadores.append(jogador)
    continuar  = input("Quer cadastrar outro amigo? (s/n) ")
    if continuar == "n":
        break #isso " quebra" o loop e o programa
# Passo 6: Mostrar o resultado final
print("\n---LISTA DE JOGADORES CADASTRADOS---")
for jogador in cadastro_jogadores:
    print(f"Nome: {jogador['nome']}, Jogo Favorito: {jogador['jogo']}")



```
